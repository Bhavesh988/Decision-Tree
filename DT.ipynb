For the given ‘Iris’ dataset, we create the Decision Tree classifier and visualize it graphically. The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.


```python
#Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
%matplotlib inline
```

```python
#Reading the datasetQ
dataset = pd.read_csv('Iris.csv' ,index_col=0)
dataset.head(5)
```

```python
#To plot multiple pairwise bivariate distributions in a dataset. 
#The pairplot() function shows the relationship for (n,2) combination of variable in a Dataframe as a matrix of plots and the diagonal plots are univariate plots.
sns.pairplot(dataset,hue="Species")
```

```python
#Independent variables
X=dataset.iloc[:,0:4]
print(X)
```

```python
#Dependent Variable
Y=dataset.iloc[:,4]
print(Y)
```

```python
#Splitting the dataset into training and test set
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size =0.3 ,random_state=0)
```

```python
#Creating a Decision Tree
classifier = DecisionTreeClassifier(criterion = 'entropy' , random_state = 0 ,max_depth=5)
classifier.fit(x_train,y_train)
```

```python
#Predicting for the test set
pred_X=classifier.predict(x_test)
pred_X
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test , pred_X)
sns.heatmap(cm,annot=True)
```

```python
#Accuracy
classifier.score(x_test,y_test)
```

```python
#Visualizing the Decision Tree
classifier = DecisionTreeClassifier(criterion = 'entropy' , random_state = 0 ,max_depth=5)
classifier.fit(X,Y)
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig = plt.figure(figsize = (20,10))
plot_tree(classifier, feature_names = fn, class_names=cn, filled = True, rounded = True)
plt.show()
```
